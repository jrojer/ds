### Корреляция

Положительная, отрицательная
Scatter plot

Ковариация
$cov = \displaystyle\frac{\sum (x_i-\bar X)(y_i-\bar Y)}{N-1}$

```
cov = sum((x-ax)*(y-ay) for x,y in zip(X,Y))/(N-1) 
```

Корреляция (коэффициент корреляции Пирсона)
$r_{xy} = \displaystyle\frac{cov}{\sigma_x \sigma_y}$  
```
rxy = cov/(sdx*sdy)
```
**важно, чтобы $X$ и $Y$ имели нормальное распределение. Критерии опирающееся на среднее значения по выборке плохо совмещаются с выбросами, ассиметрией и бимодальностью**.

Коэффициент детерминации: 
$R = r_{xy}^2$
